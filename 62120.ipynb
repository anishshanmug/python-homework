```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'state':['california','dc','california','dc','california','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],'num_pets':[5,1,0,5,2,2,3]
})
```

```python
df
```

```python
df.plot(kind='scatter',x='num_children',y='num_pets',color='blue')
plt.show()
```

```python
df.plot(kind="scatter",x="state",y="gender",color="green")

plt.show()
```

```python
df.plot(kind='bar',x='name',y='age')
plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line',x='name',y='num_children',ax=ax)
df.plot(kind='line',x='name',y='num_pets', color='red', ax=ax)

plt.show()
```

```python
df.groupby('gender')['state'].nunique().plot(kind='bar')
```

```python
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
df.groupby(['gender','state',"age"]).size().unstack().plot(kind='bar',stacked=True,logy=1)
plt.show()
```
