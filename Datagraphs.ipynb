```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Seasons_stats.csv',index_col=0)
```

```python
df
```

```python
data2 = np.arange(1,24692)
for i in range(0,24691):
    x=str(i)
    data2[i]=x
data2=data2.astype(str)
for i in range(0,24690):
    x=str("player"+ data2[i+1])
    data2[i]=x
data2[24690]="player24691"
df.index=data2.tolist()
```

```python
df.groupby(['Age','PTS']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
df.groupby(['Pos','PTS']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
df.groupby(['Age','Pos',"PTS"]).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```
